```python
import pandas as pd
import matplotlib.pyplot as plt
releaseyear=df['release_year']
duration=df['duration']
plt.bar(releaseyear,duration)
plt.title('releaseyear vs duration')
plt.xlabel('releaseyear')
plt.ylabel('duration')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
director=df['director']
releaseyear=df['release_year']
bins=[1990,1992,2000,2005,2010,2015,2020]
plt.hist(releaseyear,bins=bins,edgecolor='black')
plt.title('director vs releaseyear')
plt.xlabel('director')
plt.ylabel('releaseyear')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
country=df['country'].astype(str)
releaseyear=df['release_year'].astype(str)
plt.scatter(country,releaseyear,cmap='summer',alpha=0.75)
plt.title('country vs releaseyear')
plt.xlabel('country')
plt.ylabel('releaseyear')
plt.show()
```

```python

from wordcloud import WordCloud
country_counts = df['country'].value_counts()
wordcloud = WordCloud(max_words=100, background_color='white')
wordcloud.generate_from_frequencies(country_counts)
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Word Cloud of Countries')
plt.show()
```

```python
country_counts = df['title'].value_counts()
wordcloud = WordCloud(max_words=100, background_color='white')
wordcloud.generate_from_frequencies(country_counts)
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Word Cloud of titles')
plt.show()
```
